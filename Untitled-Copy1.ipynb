```python
import kagglehub
import pandas as pd
import holoviews as hv
from holoviews import opts

hv.extension('bokeh')

# 1. Descarga y carga de datos
path = kagglehub.dataset_download("unitednations/global-commodity-trade-statistics")
df_raw = pd.read_csv(f"{path}/commodity_trade_statistics_data.csv", nrows=50000)

# 2. Preparación limpia de datos
# Seleccionamos Exportaciones, País, Categoría y Valor
df = df_raw[df_raw['flow'] == 'Export'][['country_or_area', 'category', 'trade_usd']].copy()
df.columns = ['source', 'target', 'value']

# Agrupamos para quitar duplicados y nos quedamos con los 10 registros más grandes
# Esto garantiza que el gráfico no se sature y los índices sean correlativos
df_grouped = df.groupby(['source', 'target'])['value'].sum().reset_index()
df_final = df_grouped.nlargest(10, 'value').reset_index(drop=True)

# 3. Creación del Diagrama de Cuerdas
chord = hv.Chord(df_final)

chord = chord.opts(
    opts.Chord(
        cmap='Blues',
        edge_color='source',
        node_color='index',
        labels='source',
        title="Flujos de Comercio Global (Top 10 Exportaciones)",
        width=700,
        height=700,
        tools=['hover'],
        hover_tooltips=[
            ("Origen", "@source"),
            ("Destino", "@target"),
            ("Valor", "@value")
        ],
        node_selection_fill_color='navy',
        edge_alpha=0.8,
        node_size=15
    )
)
# 5. Guardado del archivo
hv.save(chord, 'index.html')

print("¡Hecho! El archivo index.html ha sido restaurado a la versión estable.")
chord
```
