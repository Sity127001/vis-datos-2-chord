```python
import kagglehub
import pandas as pd
import holoviews as hv
from holoviews import opts

hv.extension('bokeh')

# 1. Datos de Kaggle
path = kagglehub.dataset_download("unitednations/global-commodity-trade-statistics")
df_raw = pd.read_csv(f"{path}/commodity_trade_statistics_data.csv", nrows=10000)

# 2. Datos simplificados (Solo 10 flujos para que sea legible)
df = df_raw[df_raw['flow'] == 'Export'][['country_or_area', 'category', 'trade_usd']].copy()
df.columns = ['source', 'target', 'value']
df_grouped = df.groupby(['source', 'target'])['value'].sum().reset_index().nlargest(10, 'value')

# 3. Creación del objeto
chord = hv.Chord(df_grouped)

# 4. Estética Simplificada (Sin dimensiones conflictivas)
chord.opts(
    opts.Chord(
        cmap='Blues',
        # Usamos colores fijos o simples para evitar el KeyError
        edge_color='source', 
        node_color='index',
        
        labels='source',
        title="Flujos de Comercio Global (Datos ONU)",
        width=600, height=600,
        
        # Interactividad básica
        tools=['hover'],
        
        # Eliminamos el verde forzando el color de selección a azul
        node_selection_fill_color='navy',
        edge_selection_line_color='navy',
        node_nonselection_alpha=0.4,
        edge_nonselection_alpha=0.1
    )
)

# 5. Guardar como index.html
hv.save(chord, 'index.html')

print("¡guardado el fichero!")
chord
```
